## C-CAPM
- The following is called the **fundamental equation of asset pricing**
\begin{align*}
p_t = E_t \Big( m_{t+1}x_{t+1}\Big)
\end{align*}

where $m_{t+1}=\frac{\delta U'(c_{t+1})}{U'(c_t)}$, $\delta=$(subjective discount rate), $x_{t+1}=$(payoff)

- suppose $U(c)=\frac{c^{1-\gamma}}{1-\gamma}$, then $m_{t+1}= \delta \big(\frac{c_{t+1}}{c_t}  \big)^{-\gamma}$

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
 
data = pd.read_csv("c:/python/data/DJIYahoo.csv")
```

```python
df = pd.DataFrame()
df['Close'] = data.Close
df.index = pd.to_datetime(data.Date)
df.head()
```

```python
ret = df.pct_change()*100
ret = ret.dropna()
df['Ret']   = ret
df = df.dropna()
```

```python
plt.style.use('seaborn')
plt.title("DJIA Historical Trend and Return", fontsize=20)
df.Close.plot(figsize=(10,5))
plt.annotate('Black Monday(>-22%)', (['1987-10-01',4000]), color='red')
plt.annotate('Asian Crisis', (['1998-10-01',5500]), color='red')
plt.annotate('Another Oct. Crash', (['2002-11-01',7500]), color='red')
plt.annotate('Subprime Crisis', (['2007-10-01',5500]), color='red')
plt.annotate('Crash', (['2015-09-01',12500]), color='red')
df.Ret.plot(secondary_y=True, alpha=0.5)
plt.axhline(-10.0, c='k', ls='--')
plt.axhline(10.0, c='k', ls='--');
```
