# Fama-French 3-Factor Model(1992)

1. market risk
2. the outperformance of small versus big companies (SMB)
 - usually 30:40:30
3. the outperformance of high book/market versus small book/market companies (HML)
 - usually 30:40:30 
 - market value = price $\times$  No.Issuing. 
 - book value = accounting statement. Firm value are reflected a bit slowly

### 5-factors (2015)
- the difference between the returns of firms with robust (high) and weak (low) operating profitability
-  the difference between the returns of firms that invest conservatively and firms that invest aggressively.

\begin{align*}
(r_i - r_f) = \alpha + \beta (r_M - r_f) + \theta_s SMB + \theta_v HML +\epsilon
\end{align*}

- Results
 -  3-Factor models explained 90% of returns (CAPM:70%)
 - $\theta_s, \theta_v$: positive (significant)
   - SML: average return difference per month is 0.74%
   - HML: average return difference per month is 1.83%

### How do Fama-French to interpret SMB and HML?

- Is it against EMH view?
  - if SMB and HML are **Proxy** for unknown risk factor, them goes with EMH
  - if takes results as mispricing, then against EMH
  
  
  _"If our results are more than chance, they have practical implications for portfolio formation and performance evaluation by investors whose primary concern is long-term average returns. If asset-pricing is rational, size and BE/ME must proxy for risk”_


# De Bondt and Thaler (1985) - Does the stock market overreact?

-  연구결과 요약: 과거에 성과가 좋지 않았던 주식이 성과가 좋았 던 주식에 비해 수익률이 좋음
-  Mean Reversion(평균 회귀)
-  분석방법: 가장 최근 과거 3년 데이터로 이후 3년 동안의 수익 률을 예측가능한지 분석한 후 저자들은 분석 결과를 ‘overreaction’으로 정리함
  -  주가의 극단적인 변동 이후에는 이어서 바로 반대 방향으로의 변동 이 발생한다.
  -  초기 가격 변동이 극단적일수록, 조정폭 또한 커진다.
  
- You can make money (achieve excess return) if buying three year losers and sell three year winnder  
  
  $\Longrightarrow$  **VIOLATE WEAK EMH**


## Small firm issues

- Both papers (Fama-French and De Bondt and Thaler) use small firm data
- Conrad and Kaul (1993) attack(DT results) with arguments of 'expected return from small cap stocks has upward bias'
  - DT and other small-firm-data-used-papers' price data are 'either the bid side or offer side of the market. The true price is somewhere in b/w the bid and offer price'.
  - Let $p_t^o$ be observed price and assume that bids and offers are 20% difference from the true price

\begin{align*}
R_t^o = \frac{p_{t+1}-p_t}{p_t} =
    \begin{cases}
      \frac{p_{t+1}^o-p_t^o}{p_t^o} = \frac{1.2p_{t+1}-1.2p_t}{1.2p_t} = R_t  ~~~ \text{for offer price only}\\
      \frac{p_{t+1}^o-p_t^o}{p_t^o} = \frac{0.8p_{t+1}-0.8p_t}{0.8p_t} = R_t  ~~~ \text{for bid price only}\\
      \frac{p_{t+1}^o-p_t^o}{p_t^o} = \frac{1.2p_{t+1}-0.8p_t}{0.8p_t} = \frac{1}{2}\frac{p_{t+1}}{p_t}+ R_t  ~~~ \text{for offer and bid price at t+1 and t}\\
      \frac{p_{t+1}^o-p_t^o}{p_t^o} = \frac{0.8p_{t+1}-1.2p_t}{1.2p_t} = \frac{-1}{3}\frac{p_{t+1}}{p_t}+ R_t  ~~~ \text{for bid and offer price at t+1 and t}\\
    \end{cases}
\end{align*}
  -  Average out(all events have the same occurring probability) results: $\frac{1}{24}\frac{p_{t+1}}{p_t}+ R_t$
  
- Conrad and Kaul argued that <br>
_"the De Bondt and Thaler results no longer held up except for the month of January. January's results were not attributable to overreaction. They also noted that De Bondt and Thaler's “loser” portfolio (the portfolio that would be a winner in future periods) contained a disproportionate number of small cap stocks where the return bias using **monthly rebalancing is even worse than for larger cap stocks**."_

- Ball, Kothari, and Shanken further argue that using December-end data biased the results. <br>
_"Because the December-end evidence is likely to be biased due to the microstructure factors discussed earlier, we give greater credence to the June-end (and similar August-end) results. Whatever the source of the difference, it implies that the De Bondt and Thaler (1985, 1987) estimates of contrarian portfolio performance are **far from robust**."_


### Daniel and Titman(1997)
-  Examine stocks with high BE/ME and see if they have common risk factor
-  Results 
  -  No risk factor in common
  -  High Book to Market stocks do covary strongly not by particular risk factor but by  industrial or regional similarity
  
-  The conclusions of Daniel and Titman reinforce the idea that Fama and French's research is contrary to the EMH


# Empirics on Fama-French

## Factor construction   
[FAMA-French Library](http://mba.tuck.dartmouth.edu/pages/faculty/ken.french/data_library.html)

- SMB (Small Minus Big) is the average return on the three small portfolios minus the average return on the three big portfolios

        SMB =1/3 (Small Value + Small Neutral + Small Growth) – 
                        1/3 (Big Value + Big Neutral + Big Growth).	 

-  HML (High Minus Low) is the average return on the two value portfolios minus the average return on the two growth portfolios. 	 
         HML =1/2 (Small Value + Big Value)
                     - 1/2 (Small Growth + Big Growth).
